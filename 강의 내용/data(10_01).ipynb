```python
import matplotlib.pyplot as plt

# 날짜와 기온 데이터
dates = [
    '2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05',
    '2021-01-06', '2021-01-07', '2021-01-08', '2021-01-09', '2021-01-10'
]
min_temperature = [20.7, 17.9, 18.8, 14.6, 15.8, 15.8, 15.8, 17.4, 21.8, 20.0]
max_temperature = [34.7, 28.9, 31.8, 25.6, 28.8, 21.8, 22.8, 28.4, 30.8, 32.0]

# 객체 지향 방식으로 Figure와 Axes 생성
fig, ax = plt.subplots(figsize=(10, 5))

# Primitive Layer: 각각의 최소, 최대 기온 라인 그리기
ax.plot(dates, min_temperature, marker='o', linestyle='--', color='blue', label='Min Temperature')
ax.plot(dates, max_temperature, marker='o', linestyle='--', color='red', label='Max Temperature')

# Primitive Layer: 텍스트 추가
for i, date in enumerate(dates):
    ax.text(date, min_temperature[i] - 1, f"{min_temperature[i]}°C", ha='center', color='blue')
    ax.text(date, max_temperature[i] + 1, f"{max_temperature[i]}°C", ha='center', color='red')

# Composite Layer: 레이블, 제목, 축 설정
ax.set_title('Daily Min and Max Temperatures (Jan 2021)')
ax.set_xlabel('Date')
ax.set_ylabel('Temperature (°C)')
ax.set_xticklabels(dates, rotation=45)
ax.grid(True)

# Composite Layer: 범례
ax.legend()

# 레이아웃을 조정하고 그래프 표시
fig.tight_layout()
plt.show()
```

# matplotlib 옵션 설정

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, AutoMinorLocator, FuncFormatter
import seaborn as sns
import numpy as np

tips = sns.load_dataset("tips")
tips.head()

```

```python
tips.info() # 범주형 데이터 바람직함, object는 지양
```

```python
# seaborn

# 왼쪽 그래프
# 객체지향 선언
fig, ax =plt.subplots(nrows = 1, ncols = 2, figsize=(16, 5))
sns.barplot(data = tips,
            x = "day",
            y = "total_bill",
            errorbar=None,
            color = 'lightgray',
            alpha = 0.5,
            zorder = 2,
            ax = ax[0])

for p in ax[0].patches:
  # print("여기 확인 필요:", p, type(p))
  print("높이", p.get_height(), "너비", p.get_width())
  height = np.round(p.get_height(), 2)
  ax[0].text(p.get_x() + p.get_width()/2,
             height + 1,
             height, ha = 'center',
             size=12)


# 옵션
ax[0].set_title("Ideal Bar Graph", size = 16)
ax[0].set_ylim(-2,25) # 여백의 미
ax[0].set_xlabel("Weekday", size = 14)
ax[0].set_ylabel("Avg. Total Bill($)", size = 14)

# 오른쪽 그래프
sns.barplot(data = tips,
            x = "day",
            y = "total_bill",
            errorbar=None,
            color = 'steelblue',
            alpha = 0.9,
            zorder = 2,
            ax = ax[1])

for p in ax[0].patches:
  # print("여기 확인 필요:", p, type(p))
  print("높이", p.get_height(), "너비", p.get_width())
  height = np.round(p.get_height(), 2)
  ax[1].text(p.get_x() + p.get_width()/2,
             height + 1,
             height, ha = 'center',
             size=12)

ax[1].set_title("Just Bar Graph", size = 16)
ax[1].set_ylim(-2,25) # 여백의 미

fig.tight_layout()
plt.show()
```

```python
def major_formatter(x, pos):
    return "%.2f$" % x
formatter = FuncFormatter(major_formatter)
```

```python
group_mean = tips.groupby(['day'])['total_bill'].agg("mean")
# group_mean 가장 높은 값들
h_day = group_mean.sort_values(ascending=False).index[0]
h_mean = np.round(group_mean.sort_values(ascending=False)[0], 2)
h_day, h_mean
```

```python
# 객체 지향 선언
fig, ax =plt.subplots(nrows = 1, ncols = 2, figsize=(16, 5))
sns.barplot(data = tips,
            x = "day",
            y = "total_bill",
            errorbar=None,
            color = 'lightgray',
            alpha = 1,
            zorder = 2, # 일반적으로 2, 그래프 뒤로 보내기
            ax = ax[0])

for p in ax[0].patches:
  # print("여기 확인 필요:", p, type(p))
  # print("높이", p.get_height(), "너비", p.get_width())
 height = np.round(p.get_height(), 2)
 fontweight = "normal"
 color = "k"
 if h_mean == height:
  fontweight = "bold"
  color = "darkred"
  p.set_facecolor(color)
  p.set_edgecolor("black")

  ax[0].text(p.get_x() + p.get_width()/2,
             height + 1,
             height, ha = 'center', size=12,
             fontweight=fontweight, color=color)

# 옵션
ax[0].set_title("이렇게 그리고 싶어요!!", size = 16)
ax[0].set_ylim(-2, 30)

# spines 항목 정리
ax[0].spines['top'].set_visible(False)
ax[0].spines['left'].set_visible(False)
ax[0].spines['right'].set_visible(False)
# ax[0].spines['bottom'].set_visible(False)
ax[0].spines['left'].set_position(("outward", 20))

# y축 정리
ax[0].yaxis.set_major_locator(MultipleLocator(10)) # 큰 눈금은 10 간격으로
ax[0].yaxis.set_major_formatter(formatter) # 20.00$
ax[0].yaxis.set_minor_locator(MultipleLocator(5)) # 작은 눈금은 5 단위로

# 축 레이블 정리
ax[0].set_ylabel("Avg. Total Bill($)", size = 14)
ax[0].set_xlabel("Weekday", size = 14)

ax[0].grid(axis = 'y', which = "major", color = 'lightgray')
ax[0].grid(axis = 'y', which = "minor", linestyle = ':')

# 오른쪽 그래프
sns.barplot(data = tips,
            x = "day",
            y = "total_bill",
            errorbar=None,
            alpha = 0.5,
            zorder = 2,
            ax = ax[1])

for p in ax[1].patches:
  # print("여기 확인 필요:", p, type(p))
  # print("높이", p.get_height(), "너비", p.get_width())
  height = np.round(p.get_height(), 2)
  ax[0].text(p.get_x() + p.get_width()/2, height + 1, height, ha = 'center', size=12)

# 옵션
ax[1].set_title("Sample", size = 16)
ax[1].set_ylim(-2, 25)

plt.show()
```
